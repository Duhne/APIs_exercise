```python
# import the SDK and instantiate the crypto historical data client
from alpaca.data.historical import CryptoHistoricalDataClient

# No keys required for crypto data
client = CryptoHistoricalDataClient()
```

```python
# define the parameters for our request
#  Import the request class for crypto bars, CryptoBarsRequest and TimeFrame class to access time frame units more easily.
from alpaca.data.requests import CryptoBarsRequest
from alpaca.data.timeframe import TimeFrame

```

```python
# Creating request object
request_params = CryptoBarsRequest(
                        symbol_or_symbols=["BTC/USD"],
                        timeframe=TimeFrame.Day,
                        start="2022-09-01",
                        end="2022-09-07"
                        )
```

```python
# Retrieve daily bars for Bitcoin in a DataFrame and printing it
btc_bars = client.get_crypto_bars(request_params)

# Convert to dataframe
btc_bars.df
```

```python
# live 
from alpaca.data.live import CryptoDataStream
```

```python
crypto_stream = CryptoDataStream(apikey_crypto, secret_key_crypto)
```

```python
async def bar_callback(bar):
    for property_name, value in bar:
        print(f"\"{property_name}\": {value}")

# Subscribing to bar event 
symbol = "BTC/USD"
crypto_stream.subscribe_bars(bar_callback, symbol)

crypto_stream.run()
```

```python
def flatten(data_disney, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data_disney[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data_disney = pd.concat([data_disney, flattened], axis=1)
        data_disney = data_disney.drop(column, axis=1)
    return data_disney
```
